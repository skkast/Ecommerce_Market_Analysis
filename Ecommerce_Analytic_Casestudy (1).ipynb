### c. Understand the retention of customers on month on month basis

```python
Order_data
```

```python
# Get list of customers per month
customers_by_month = Order_data.groupby('Purchase_month')['customer_id'].unique()

# Calculate how many customers were retained each month
retention_counts = []

months = sorted(customers_by_month.index)

for i in range(len(months) - 1):
    current_month = months[i]
    next_month = months[i + 1]
    
    current_customers = set(customers_by_month[current_month])
    next_customers = set(customers_by_month[next_month])
    
    retained_customers = current_customers & next_customers  # intersection
    
    retention_counts.append({
        'Month': str(next_month),
        'Retained_Customers': len(retained_customers)
    })

# Convert to DataFrame
retention_df = pd.DataFrame(retention_counts)
print(retention_df)
```

## 3. Cross-Selling (Which products are selling together)
## Hint: We need to find which of the top 10 combinations of products are selling together in
## each transaction. (combination of 2 or 3 buying together)

```python
order_products = Order_merge.groupby('order_id')['product_id'].apply(list)
```

```python
order_products
```

```python
from itertools import combinations
from collections import Counter

# Store all combinations
combo_counts = Counter()

# Loop through orders
for products in order_products:
    # Remove duplicates in case product is repeated in same order
    products = list(set(products))
    
    # Generate 2-product and 3-product combos
    for r in [2, 3]:
        combos = combinations(products, r)
        combo_counts.update(combos)
```

```python
top_10_combos = combo_counts.most_common(10)

# Display
for combo, count in top_10_combos:
    print(f"{combo} â†’ bought together {count} times")
```
